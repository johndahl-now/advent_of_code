```python
with open('input.txt','r') as fh:
    data = fh.readlines()
    data = ''.join(data).split('\n')
```

```python
data
```

```python
rows = len(data)
cols = len(data[0])
slopes = (
    (1,1),
    (3,1),
    (5,1),
    (7,1),
    (1,2)
)
```

```python
def count_trees(slope):
    x, y, trees = 0, 0, 0
    while(True):
        try:
            col = x if x < cols else x % cols
            if data[y][col] == '#':
                trees += 1
        except IndexError:
            break
        x += slope[0]
        y += slope[1]
    return trees
```

```python
count_trees(slopes[1])
```

```python
trees = []
for slope in slopes:
    trees.append(count_trees(slope))
```

```python
trees
```

```python
answer = 1
for tree in trees:
    answer *= tree
print(answer)
```
